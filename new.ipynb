```python
#Open website

def new_windown():
    options = webdriver.EdgeOptions()
    driver = webdriver.Edge(options=options)
    driver.get("https://www.google.com.br/")
    wait = WebDriverWait(driver, 3)
    # driver.maximize_window()
    # wait - WebDriverWait(driver, 3)
    return driver
```

```python
#Choose which products you want to compare
option1 = ['Iphone 15 Pro Max','S24 Ultra']
option2 = ['Notebook Gamer Legion Slim 5i (16" Intel)','Notebook Gamer G15']
choose = (input(f'Which products do you want to compare? Option 1:{option1} Option 2: {option2}'))
print(choose)
# option1 = 1
# option2 = 2
```

```python
# Listas vazias para serem iteradas
list_links = []
list_links2 = []

if (choose == "1"):
    # Abre uma nova guia
    driver = new_windown()

    search = driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea")
    search.send_keys(option1[0])
    search.send_keys(Keys.RETURN)

    sleep(10)

    html = driver.page_source

    # Cria um objeto BeautifulSoup para analisar o HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Obter a URL atual
    url = driver.current_url

    # Fecha o driver do Selenium
    driver.quit()

    # Faz uma solicitação GET para obter o conteúdo HTML da página
    response = requests.get(url)

    # Verifica se a solicitação foi bem-sucedida (código de status 200)
    if response.status_code == 200:
        # Cria um novo objeto BeautifulSoup para analisar o HTML da página
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrando todos os links dos resultados da pesquisa
        links = soup.find_all('a')

        # Iterando sobre a lista e extraindo apenas os URLs
        for link in links:
            href = link.get('href')
            if href.startswith('/url?q='):
                # Extraindo o URL do link
                url = href.split('/url?q=')[1].split('&')[0]
                list_links.append(url)
    print(list_links[2])

    # Abre uma nova guia
    driver = new_windown()

    search = driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea")
    search.send_keys(option1[1])
    search.send_keys(Keys.RETURN)

    sleep(10)

    html = driver.page_source

    # Cria um objeto BeautifulSoup para analisar o HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Obter a URL atual
    url = driver.current_url

    # Fecha o driver do Selenium
    driver.quit()

    # Faz uma solicitação GET para obter o conteúdo HTML da página
    response = requests.get(url)

    # Verifica se a solicitação foi bem-sucedida (código de status 200)
    if response.status_code == 200:
        # Cria um novo objeto BeautifulSoup para analisar o HTML da página
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrando todos os links dos resultados da pesquisa
        links = soup.find_all('a')

        # Iterando sobre a lista e extraindo apenas os URLs
        for link in links:
            href = link.get('href')
            if href.startswith('/url?q='):
                # Extraindo o URL do link
                url = href.split('/url?q=')[1].split('&')[0]
                list_links2.append(url)
    print(list_links2[2])
```
